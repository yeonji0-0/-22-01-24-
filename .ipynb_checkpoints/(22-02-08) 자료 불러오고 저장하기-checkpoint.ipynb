```python
data = pd.read_csv(path, sep = ",")
data.head()
```

```python
#unnamed0 컬럼을 id_area컬럼 이름 변경  -> rename
data.rename(columns = {"Unnamed: 0" : "id_area"}, inplace=True)  #대소문자 공백 체크!
data.head()
```

```python
data.columns[0]
```

```python
#대소문자 공백 신경쓰고 싶지 않다면
data.rename(columns = {data.columns[0] : "id_area"}, inplace=True)  
data.head()
```

```python
#id_area컬럼의 이름을 지정했는데 진짜 인덱스로 쓸 만한지 체크
# -> id_area값이 다 유니크한지 체크
data["id_area"]   #FM방법 : data.loc[ :, "id_area"]
len(data["id_area"].unique())
```

```python
data.shape[0] == len(data["id_area"].unique())
```

```python
data.head()
```

```python
#step1) 내가 하고자 하는 것에 대한 sample을 가지고 구현
str(int("4.North-Apulia".split(".")[0])-1)+"_"+"4.North-Apulia".split(".")[1]
```

```python
#step2) 위의 step1에서 무엇이 입력 x로 바뀔지
# -> lambda 함수로 어떻게 표현할지 고민 + 모든 가로줄에 다 태워야 하니 apply
data["id_area"] = data["id_area"].apply(lambda x : str(int(x.split(".")[0])-1)+"_"+x.split(".")[1] )
data.head()
```

```python
# palmitoleic  컬럼 값들을 75 -> 75/100, 57 -> 57/100 소수값들로 바꾸기
data["palmitoleic"] = data["palmitoleic"]/100
data.head()
```

```python
data["palmitoleic"].apply(lambda x : x/100)
```

```python
#기존 컬럼을 -> 가로줄에 대한 인덱스로 변환을 할 때
#set_index
```

```python
data.index
```

```python
data.set_index("id_area", inplace=True)
data.head()
```

```python
#작업한 DataFrame을 외부 파일로 출력할 때 to_양식
data.to_csv("01_data/testcsv.csv")
```
