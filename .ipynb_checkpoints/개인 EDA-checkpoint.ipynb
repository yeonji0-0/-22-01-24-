```python
### 모듈 import
import xml.etree.ElementTree as ET

### 압축파일 안의 xml 파일 읽기
tree = ET.parse('corpCode.xml')
root = tree.getroot()
```

```python
### 회사 이름으로 회사 고유번호 찾기
def find_corp_num(find_name):
    for country in root.iter("list"):
        if country.findtext("corp_name") == find_name:
            return country.findtext("corp_code")
```
