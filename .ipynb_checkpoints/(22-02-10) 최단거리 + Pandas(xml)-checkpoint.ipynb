### 다익스트라 알고리즘 그냥 구현하는 방식

```python
#입력단 설계
INF = float("inf")

#노드의 수 n(도시의 수), 연결 도로의 수 m
n,m = map(int, input("도시 수, 연결 도로수 공란 입력").split())
#시작 도시 정보
start = int(input("시작 도시 입력"))
#도로에 대한 정보 입력 : 출발, 도착, 둘 사이 거리 -> 3가지 정보
graph = [ [] for _ in range(n+1)] # -> 1번 도시 정보를 graph의 1번 인덱스와 연결
for _ in range(m):
    s, e, d = map(int, input("S,E,Dis").split())
    graph[s].append( (e,d)) #S도시에서 종착점 거리까지 거리 묶어서
graph
```

```python
graph[1][0]
```

```python
# 필요한 것 세팅
distance = [INF] * (n+1) # 0번 도시~ n번 도시까지 맵핑하려고
# 방문했던 도시인지 체크
visited  = [False] * (n+1)
print(distance)
print(visited)
```

```python
# 다음 도시를 어디로 가야할지~ : 함수!!!
# 입력 : 그냥 없음!!!
# 출력 : 방문하지 않은 도시 중 거리가 제일 짧은 친구
def get_smallest_node():
    # 가장 거리가 짧은 거리 & 도시 번호
    min_value = INF
    min_index = 0 # -> 도시별로 돌아가면서, visited False 인 것들에서 거리 짧은 것을 min_value에 기록하면서

    # 각 도시를 다 돌리면서..
    for node in range(1, n+1) : # 시작을1로 하는 이유는 0번 도시 pass~ n번 도시까지 다 
        if ( not visited[node]) and (distance[node] < min_value) :
            min_value = distance[node]
            min_index = node
        else:
            pass
    return min_index
```

```python
## 본격적인 최단거리 탐색함수
#  함수 입력 : 출발도시
#  출력      : 딱히 없고,,,외부에 있는  distance에 접근해서 갱신하도록
def dijkstra(start):
    # 시작점에 대해서 처리~ : 거리 & 방문도장
    distance[start] = 0
    visited[start] = True
    
    # 출발 도시에 대해서...연결 도시의 정보를 가지고 와야 함
    for j in graph[start]: # -> j : j[0] 종착점, j[1] 둘 사이 거리 
        distance[j[0]] = 0 + j[1]
    #############################################################
    # 출발 도시는 끝났으니
    # 남은 도시들을 꺼내면서 어디로 갈지
    for _ in range(n-1): # 이유는 시작도시는 끝났기에
        # 지금 제일 거리가 짧은 안 간 도시는 누구지?
        now = get_smallest_node()
        visited[now] = True
        # 그 도시에서 연결된 애들,,,
        for k in graph[now]:
            new_distance = distance[now] + k[1]
            # 거리를 업데이트를 할지 말지에 대한 판단
            if new_distance < distance[k[0]]:
                distance[k[0]] = new_distance
            else:
                pass
    
```

### 다익스트라 알고리즘 heqpq 이용해서 수정 

```python
# 입력단 설계 #
INF = float("inf")

n,m,k,x = list(map(int, input("도시수,도로수,거리,시작도시").split()))
print(n,m,k,x)
# 시작 도시 정보
start = x

# 도로에 대한 정보 입력 : 출발, 도착, 둘 사이 거리 -> 3가지 정보
graph = [ [] for _ in range(n+1)] # -> 1번 도시 정보를  graph의 1번 인덱스와 연결
for _ in range(m):
    s, e = map(int, input("S,E").split() )
    graph[s].append( (e, 1) ) # S도시에서 종착점과 거기까지 거리 묶어서~~
graph
```

```python
import heapq
```

```python
# 거리에 대한 정보 초기화
distance = [INF] * (n+1)

# 최단거리 탐색
def dijk_city(start):
    # 처음 도시에 대해서 거리 처리 & 큐에 넣기(거리, 도시) -> 순서 주의
    distance[start] = 0
    q = [] 
    heapq.heappush(q, (0, start))

    ### 큐가 빌 때 까지
    while q:
        dist, now = heapq.heappop(q)
        if distance[now] < dist:
            continue
        else:
            for j in graph[now]: #j는 입력시 종착도시, 거리
                cost = dist + j[1] # now를 거쳐 j[0]도시 갈 때 예상거리
                if cost < distance[j[0]]:
                    distance[j[0]] = cost
                    heapq.heappush(q, (cost, j[0]))
```

### 카카오 택시 문제

```python
#먼저 필요한 패키지 설정
import heapq
#필요한 기본 값 설정 :INF
INF = float("inf")
```

```python
def solution(n, s, a, b, fares):
    answer = 0
    ############################
    ### 1) 입력부분 처리
    #   1-1) 지도,거리 : 0번하고, 1번~n번까지 사용
    graph = [ [] for _ in range(n+1)]
    distance = [INF for _ in range(n+1)]
    #   1-2) 지도 내의 도로와 요금 입력
    for fare in fares: # fare : 도시1, 도시2, 둘 사이 요금
        # 양방향으로 길을 구성해야하니...
        # 도시1출발, 도시2도착, 요금
        # 도시2출발, 도시1도착, 요금
        s_node,e_node,f = fare
        graph[s_node].append((e_node,f))
        graph[e_node].append((s_node,f))
        # print(fare, s,e,f)
        # print(fare,fare[0], fare[1],fare[2])
        # graph[fare[0]].append((fare[1],fare[2]))
        # graph[fare[1]].append((fare[0],fare[2]))
    ### 2) 주어진 입력 s에서 출발할 떄, 모든 도시에 이르는 최소 요금 계산
    #      -> dijkstra
    distance = dijkstra(s, distance, graph)
    # -> distance 변수에는 입력start에서 각 도시별 최소 요금이 기록!!

    ### 3) 문제에서 출발점에서의 합승할 때 각 지점별 최소 금액은 계산했으니
    #   -> 도시별로 출발해서, 각 도시별로 최소 요금을 구하고
    #   -> 빨간색 + 파란색(a,b) :전체 요금-> 최소..
    min_fare = INF
    for start_node in range(1, n+1):
        # 이제 각 도시별로 이제 Bye Bye해서 간다할 때, 모든 도시까지 최소 요금 기록
        dist = [INF for _ in range(n+1)]
        dist = dijkstra(start_node,  dist, graph)
        min_fare = min(min_fare, distance[start_node] + dist[a] + dist[b])
    answer = min_fare
    ############################
    return answer
```

```python
# 실제 시작점에서 남은 도시들까지 최단요금/거리 계산하는 함수..
# 입력 : 출발점, 거리 기록지, 지도
# 출력 : 출발점에서 갈 때 각 도시별 최소요금
def dijkstra(start, distance, graph):
    # 일단 필요한 우선순위 큐
    q = []
    # 시작점에 대해서 처리(거리0세팅, 큐에 넣기)
    distance[start] = 0
    heapq.heappush(q, (0, start)) # 거리, 도시 순으로 넣기
    # 열심히 우선순위큐가 빌때 까지 가기
    while q:
        # 일단 어디로 갈지 뽑기
        dist, now = heapq.heappop(q)
        # 이미 왔던 곳인지 체크
        if distance[now] < dist:
            continue
        else: # 처음 오는 곳이니 값 갱신
            for i in graph[now]: # 도시, 요금
                cost = dist + i[1]
                if cost < distance[i[0]]:
                    distance[i[0]] = cost
                    heapq.heappush(q, (cost, i[0])) # 우선순위 큐 : 요금 ,도시
    return distance
```

### 화성탐사 문제

```python
import heapq
INF = float("inf")
```

```python
dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]
```

```python
test_num = int(input("테스트를 할 횟수를 입력하세요 "))

for tc in range(test_num):
    print(tc)
    # Test case의 지도의 크기!!!!
    n = int(input("지도 크기 입력해요 "))
    # 지도 생성
    graph = [] # 아래는 지도의 입력을 받는 부분~!!!!!
    for i in range(n):
        ### 가로줄의 정보 받아~~~-> 지도!!!!
        # input("{0}번째 가로줄 입력하세요".format(i+1)) # "1 3 5"
        # input(~~).split() "1","3","5"
        # map(int, input().split()) 1,3,5
        # list( map(int, input().split())) [1, 3, 5]
        graph.append(list(map(int, input("{0}번째 가로줄 입력 ".format(i+1)).split())))
    ## 지도 입력 되었으니~~초기 세팅!!!!! -> 시작점의 위치, 거리 세팅

    ### 초기화~~~~~~
    # 거리정보에 대한 것 : 이 문젣에서는 n by n으로
    distance = [ [INF] * n for _ in range(n)]
    # 움직이는 시작점!!!! -> 2차원에서이니, x, y로하고 ,시작점 
    x, y = 0, 0
    # 시작점에서 나 왔다!!! -> 도장!!!!! 거리
    distance[x][y] = graph[x][y]
    # 시작점에 대한 정보를 -> 큐~~ : 우선순위퀴(값, ~~) -> 값인 작은거~~
    q = [ (graph[x][y], x,y)]
    # 큐가 빌 때 까지~~~
    while q:
        # 이제 움직일 곳은 뽑자~~~ : 거리가 제일 작은 친구가 나오는데ㅔ
        # 정보 : 거리, 가로, 세로
        dist, x, y = heapq.heappop(q)
        # 처리할 필요가 없는 경우
        if distance[x][y] < dist:
            continue
        # 처리해야할 경우~~ : 상, 하 , 좌, 우
        for i in range(len(dx)):
            nx = x + dx[i]
            ny = y + dy[i]
            # 이동이 가능한지 : 안쪽, 바깥쪽
            if nx <0 or nx >= n or ny <0 or ny >= n:
                continue
            else: # 이동이 가능한 경우....
                cost = dist + graph[nx][ny]
                # 갱신을 할지 말지~~
                if cost < distance[nx][ny]:
                    distance[nx][ny] = cost
                    heapq.heappush(q, (cost, nx,ny))
    print(distance[n-1][n-1])  # tc가 끝날 때 마다
```
