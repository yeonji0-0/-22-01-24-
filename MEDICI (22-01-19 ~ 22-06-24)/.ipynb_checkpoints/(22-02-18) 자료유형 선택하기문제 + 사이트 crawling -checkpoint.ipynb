### Hash table (자료유형 선택하기)

#### 마라톤 문제

```python
#내가 푼 방식
data = []

def sol(participant,completion):
    for i in range(len(participant)):    #이중 for문은 효율성 문제 발생할 수 있음
        for j in range(len(completion)):
            if participant[i] == completion[j]:
                data.append(participant[i])
    reslut = set(participant) - set(data) # 동명이인이 있기 때문에 set으로 처리할 수 없음
    return reslut
```

```python
participant_1 = ["leo", "kiki", "eden"]
completion_1 = ["eden", "kiki"]
sol(participant_1,completion_1 )
```

```python
def solution(participant,completion):
    answer = ''
    #################
    #참가자(n) -> 이름 : key, 명수 : value -> 1 for
    people_dict = {}
    for p in participant:
        if p in people_dict: #이미 참가자 명단에 있다면
            people_dict[p] += 1
        else: #이름이 처음인 사람
            people_dict[p] = 1
            
    #완주자(n) -> 참가자dict에서... key -> 접근, 인원수를 -1 차감
    #### 1) 1명 참가, 1명이 완주 : key 삭제
    #### 2) 동명이인들, 그 중 1명이 완주 : value -1
    for p in completion:
        if people_dict[p] == 1: #그 사람 이름으로 1명 존재 & 완주 했대
            del people_dict[p]
        else: #완주는 했는데, 여러 있는 경우
            people_dict[p] -= 1
            
    answer = list(people_dict.keys())
    
    ################
    return answer
```

#### 스파이 문제

```python
def solution(clothes):
    answer = 0
    ################################
    cloth_dict = {} # 키 : 종류, 이름 : 구체적인 의상들
    # 주어진 입력은 (이름, 종류)
    for name, category in clothes:
        # category가 new -> 키로 생성
        # category가 unnew -> 값에 추가
        if category not in cloth_dict: # dict-> 키에 대해서 체크
            cloth_dict[category] = [name] # 동일category에 다른 name와도 리스트로 담을려고
        else: # 이미 기존에 category있으니..-> value 값만 추가
            cloth_dict[category].append(name)
    ############# 입력들에 대한 dict 처리 완료

    # 계산 : 항목별 수(+1) * 항목별 수(+1) .... -1
    # 곱의 계산을 위해서... 
    answer = 1
    for k, v in cloth_dict.items():
        answer *= (len(v)+1)
    answer -= 1
    ################################
    return answer
```

```python
clothes = [["yellowhat", "headgear"], ["bluesunglasses", "eyewear"], ["green_turban", "headgear"]]
solution(clothes)
```
