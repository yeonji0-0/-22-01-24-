```python
#Q) 5번째 영화의 코드값 : movieCd
#Q) 5번째 영화의 국문명 : movieNm
#Q) 5번째 영화의 영문명 : movieNmEn
#Q) 5번째 영화의 개봉일 : openDt
#Q) 5번째 영화의 감독이 있다면 첫번째 감독 : directors
idx = 5
temp = movie_data["movieListResult"]["movieList"][idx]
print(temp["movieCd"])
print(temp["movieNm"])
print(temp["movieNmEn"])
print(temp["openDt"])
if temp["directors"] == []:
    print("")
else:
    print(temp["directors"][0]["peopleNm"])
```

```python
#아무것도 없는 빈 상태에서 정보를 추출해서 누적하면서 데이터를 쌓으려면?
#일단 필요한 DF : 2개  -> 1개 영화 정보 담을 DF : Only 1st / 누적할 DF : 0층부터 ??층까지
movie_df = pd.DataFrame()
movie_df_temp = pd.DataFrame()
movie_df_temp = movie_df_temp.append({"movieCd": "","movieNm":"","movieEn":"","openDt":"","DirName":""}, ignore_index=True)

movie_df_temp
```

```python
#내가 받은 영화의 수를 기반으로 해야하니
num = len(movie_data["movieListResult"]["movieList"])
temp_code_id = movie_df_temp.columns.get_loc("movieCd") #컬럼 이름을 바로 쓰지 않고 위치로 추출해서 쓰면 오류 날 일 줄어듦
temp_name_id = movie_df_temp.columns.get_loc("movieNm") 
temp_enname_id = movie_df_temp.columns.get_loc("movieEn") 
temp_open_id = movie_df_temp.columns.get_loc("openDt") 
temp_dirname_id = movie_df_temp.columns.get_loc("DirName") 
#i번째 영화에 대해서 정보를 추출..(정수)
for i in range(0,num):
    #i번째 영화에 대해서 필요한 정보(코드,이름, 개봉일..) 추출 -> 단, 정보가 없을 수도 있으니 예외처리 필요
    i_movie = movie_data["movieListResult"]["movieList"][i]
    # -> temo에 기록 (가로:i번째, 세로: 그 정보 컬럼 정수)
    movie_df_temp.iat[0,temp_code_id] = i_movie["movieCd"]
    movie_df_temp.iat[0,temp_name_id] = i_movie["movieNm"]
    movie_df_temp.iat[0,temp_enname_id] = i_movie["movieNmEn"]
    movie_df_temp.iat[0,temp_open_id] = i_movie["openDt"]
    if i_movie["directors"] !=[]: # 감독정보 있을 때.
        movie_df_temp.iat[0,temp_dirname_id] = i_movie["directors"][0]["peopleNm"]
        
    #temp것을 movie_df에 추가
    movie_df = movie_df.append(movie_df_temp,ignore_index=True)
    
    #temp싹 방정리 -> 다음 영화 정보를...
    for j in range(len(movie_df_temp.columns)):
        movie_df_temp.iat[0, j] = ""
    
```

```python
movie_df
```

```python
#openDt를 year - month - day로 구분해주기
movie_df["year"] = movie_df.iloc[ :,temp_open_id].apply(lambda x : str(x)[:4] )
movie_df["month"] = movie_df.iloc[ :,temp_open_id].apply(lambda x : str(x)[4:6] )
movie_df["day"] = movie_df.iloc[ :,temp_open_id].apply(lambda x : str(x)[6:] )
movie_df
```

```python
#컬럼을 가로줄에 대한 내가 만든 인덱스로
#일단 하기전에 openDt 순서대로 먼저 정렬하고 그 순서대로 가로줄 인덱스로
movie_df.sort_values(by = "openDt", ascending = False, inplace = True)
#컬럼을 주고, 그 컬럼을 가로줄 인덱스로 지정
#주의) 그 컬럼이 없으면 에러
movie_df.set_index("movieCd", inplace=True)
movie_df
```

```python
#컬럼 순서를 변경
movie_df = movie_df.loc[ :, ["openDt","year","month","day","movieNm","movieEn"]]
movie_df.head()
```

```python
#잠시) 상세 정보 요청하고, 처리하는 기능 체크
#1) 광해에 대한 상세 정보 요청 url (본인의 키값으로 작성)
url_p1 = "https://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key="
key = "955ef5d888fc11ea84bbe892453318e3"
url_p2 = "&movieCd="
movie_code = "20124079" #영화 코드만 변경하면 영화 바뀜
url = url_p1 + key + url_p2 + movie_code
#2) 그 주소에 대하여 통신(urllib.request.urlopen)
temp_page = urllib.request.urlopen(url)
#3) 파이썬에서 접근이 용이하게  json패기지써서 문자열 데이터를 파이썬dict, list처럼 변형
movie_info = json.loads(temp_page.read().decode("utf-8"))
movie_info
```

```python
#Q) 광해 영화의 장르는?
print(movie_info["movieInfoResult"]["movieInfo"]["genres"][0]["genreNm"])
#Q) 광해 영화의 출연한 배우의 수는?
print(len(movie_info["movieInfoResult"]["movieInfo"]["actors"]))
#Q) 광해 영화의 배우 정보가 있다면, 1번 배우에 대한 이름
print(movie_info["movieInfoResult"]["movieInfo"]["actors"][0]["peopleNm"])
#Q)                                    "  "                의 배역 이름
print(movie_info["movieInfoResult"]["movieInfo"]["actors"][0]["cast"])
```

```python
#movie_df의 기존 DF에서 장르 컬럼 추가 + 출연 배우 수 추가  +  1번 배우 배역 이름 추가 + 1번 배우 배역 이름 추가 -> 옆으로 확장
movie_df["Genre"] = ""
movie_df["ActNum"] = ""
movie_df["actActName"] = ""
movie_df["ActCast"] = ""

movie_df
```

```python
#접근 1) movie_df를 대상으로 영화코드값을 20개 출력해보세요
#    단, movie_df에서 반복문 for를 중심으로

for i_code in movie_df.index:
    #print(i_code)
    # i_code에 대한 상세정보를 요청하는 url만들기
    url_p1 = "https://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key="
    key = "955ef5d888fc11ea84bbe892453318e3"
    url_p2 = "&movieCd="
    movie_code = i_code
    url = url_p1 + key + url_p2 + movie_code
    #접근 2) 영화코드별로 상세정보를 요청하는 url을 출력해보세요
    #print(url)
    # i_code에 대한 상세정보 url을 직접 요청 -> 받아서 -> 처리하기 편하게 json 
    temp_page = urllib.request.urlopen(url)
    temp_data = json.loads(temp_page.read().decode("utf-8"))
    #필요정보 추출 -> DF에 위치에 맞게 업데이트
     # 1) 배우 수
    movie_df.at[i_code, "ActNum"] = len(temp_data["movieInfoResult"]["movieInfo"]["actors"])
    # 2) 장르 : ???혹시 장르 정보가 비면...에러가 날 텐데...
    movie_df.at[i_code, "Genre"] = temp_data["movieInfoResult"]["movieInfo"]["genres"][0]["genreNm"]
    # 3) 1번 배우에 대한 정보...배우 정보가 있을 때와 없을때..
    if len(temp_data["movieInfoResult"]["movieInfo"]["actors"]) < 1:
        movie_df.at[i_code, "ActName"] = ""
        movie_df.at[i_code, "ActCast"] = ""
    else: # 배우 정보가 1명 이상있다는 것이니...
        first_act = temp_data["movieInfoResult"]["movieInfo"]["actors"][0]
        movie_df.at[i_code, "ActName"] = first_act["peopleNm"]
        movie_df.at[i_code, "ActCast"] = first_act["cast"]
movie_df
```

```python
# 상세정보 요청하고 하는 큰 틀을 정수인덱스로!!!
temp_genre_id = movie_df.columns.get_loc("Genre")
temp_actnum_id = movie_df.columns.get_loc("ActNum")
temp_actname_id = movie_df.columns.get_loc("actActName")
temp_castname_id = movie_df.columns.get_loc("ActCast")
for i in range(len(movie_df)):
    url_p1 = "https://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key="
    key = "955ef5d888fc11ea84bbe892453318e3"
    url_p2 = "&movieCd="
    url = url_p1 + key + url_p2 + movie_df.index[i]
    temp_page = urllib.request.urlopen(url)
    temp_data = json.loads(temp_page.read().decode("utf-8"))
    
    movie_df.iat[0,temp_genre_id] = temp_data["movieInfoResult"]["movieInfo"]["genres"][0]["genreNm"]
    movie_df.iat[0,temp_actnum_id] = len(temp_data["movieInfoResult"]["movieInfo"]["actors"])
     if len(temp_data["movieInfoResult"]["movieInfo"]["actors"]) < 1:
         movie_df.iat[0,temp_actname_id]] = ""
         movie_df.iat[0,temp_castname_id] = ""
    else: # 배우 정보가 1명 이상있다는 것이니...
        first_act = temp_data["movieInfoResult"]["movieInfo"]["actors"][0]
        movie_df.iat[0,temp_actname_id] = first_act["peopleNm"]
         movie_df.iat[0,temp_castname_id] = first_act["cast"]
    
movie_df
```
