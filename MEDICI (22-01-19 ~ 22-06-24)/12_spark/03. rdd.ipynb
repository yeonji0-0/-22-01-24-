```python
from pyspark import SparkContext
```

```python
sc = SparkContext('local')
```

```python
dt = [(1,2),(3,4),(5,6),(1,7),(3,4)]
nRdd = sc.parallelize(dt)
nRdd.collect()   #앞에 있는 값을 key로 보고 뒤에 있는 값을 value로 봄
```

```python
# transfomation
rdd1 = nRdd.reduceByKey( lambda a,b : a+b)
rdd1.collect()
```

```python
nRdd.flatMap( lambda v: v).collect()
```

```python
rdd2 = nRdd.flatMap( lambda v: v).map(lambda v: (v,1))
rdd2.collect()
```

```python
rdd2.reduceByKey( lambda a,b: a+b).collect()
```

```python
# action function
rdd2.countByKey()
```

```python
#transform
rdd2.sortByKey().collect()
```

```python
rdd2.sortByKey(ascending = False).collect()
```

```python
nRdd.sortBy(lambda v: v[1]).collect()
```

## 연습문제

#### 1. k.txt에 있는 각 단어의 개수를 구하시오

```python
rddTxt = sc.textFile('k.txt')
rddTxt.collect()
```

```python
word = rddTxt.flatMap(lambda v : v.split()).map(lambda v: (v,1))
word.countByKey()
```

```python
# 정렬
word_count = word.reduceByKey(lambda a,b:a+b).sortBy(lambda v:v[1], ascending=False)
word_count.collect()
```
