```python
import pandas as pd
```

```python
baseball_player = pd.DataFrame(
    {
        "player_name" : ["권혁","이범호","필","박병호","양준혁"],
        "team_no" : [1993, 2001, 2001, 2010, None]
    }
)
baseball_player
```

```python
baseball_team = pd.DataFrame(
    {
        "team_name" : ["한화","두산","기아","넥센"],
        "team_home": ["대전","서울","광주","고척"],
        "team_id" : [1993,1999,2001,2010]
    }
)
baseball_team
```

```python
#선수 정보를 중심으로 진행을 하자
#-> 선수 정보는 팀 정보가 없어도 다 있으면 좋겠다
#-> 기준 DF : 선수 DF (L)
#-> 참조DF : 팀 DF (R)
#--- 연결 : 팀의 고유값 (선수 DF : 컬럼 team_no) (팀 DF : 컬럼 team_id)
#--> 세팅을 선수 L, 팀 R : Left Join
#--> 세팅을 팀 L, 선수 R : Right Join

#선수 L, 팀 R : Left Join
#-> 기분이 컬럼이면 on쪽, 기준이 index -> index.
pd.merge(baseball_player, baseballteam, how = "left", left_on="team_no", right_on = "team_id" )
```

```python
#팀 정보는 다 있으면 좋겠다 (변경되거나, 없어진 팀도 보고싶다)
#-> 참조 DF : 선수 DF (L)
#-> 기준DF : 팀DF (R)
#--- 연결 : 팀의 고유값 (선수 DF : 컬럼 team_no)  (팀 DF : 컬럼 team_id)
#선수 L, 팀 R : Right Join
#-> 기분이 컬럼이면 on쪽, 기준이 index -> index.
pd.merge(baseball_player, baseballteam, how = "right", left_on="team_no", right_on = "team_id" )
```

```python
#동일한 세팅(L, R)에서 지금 KBO리그에 참여하는 팀이면서, 선수들만 보고싶다
#선수 L, 팀 R : Right Join
#--- 연결 : 팀의 고유값 (선수 DF : 컬럼 team_no)  (팀 DF : 컬럼 team_id)
#----inner join, cross join
pd.merge(baseball_player, baseball_team, how ="inner", left_on="team_no", right_on = "team_id")
```

```python
#선수 정보는 있는데, 팀 정보가 없는 선수들을 보고자 함
#-> 팀 정보가 실수로 누락 됐는지, 아니면 은퇴를 한 선수들인지, 아니면 다른 나라 리그에서  하는 선수인건지 등등
temp = pd.merge(baseball_player, baseball_team, how ="left", left_on="team_no", right_on = "team_id")
temp[ temp["team_id"].isnull() ] #팀에 대한 정보가 없는 애들만 보여줘 -> 조건 검색 -> 불리언 인덱싱
```

```python
temp = pd.merge(baseball_player, baseball_team, how ="right", left_on="team_no", right_on = "team_id")
temp[ temp["player_name"].isnull() ]
```

```python
#선수 정보도 다 보고, 팀 정보도 다 보자
#중간에 한 쪽 빠져도 일단 다 합쳐서 봅시다
# full join, outer join
pd.merge(baseball_player, baseball_team, how = "outer", left_on="team_no", right_on="team_id")
```

```python
#위의 outer join을 기반으로 혹시 정보가 한쪽으로라도 누락된 데이터가 뭐지?
temp = pd.merge(baseball_player, baseball_team, how ="outer", left_on="team_no", right_on = "team_id")
temp[ (temp["player_name"].isnull())  | (temp["team_id"].isnull())]
```

```python
#위의 DF에서 좀 다르게 기준이 
#1개는 컬럼에, 1개는 가로줄 인덱스에
```

```python
baseball_player1 = pd.DataFrame(
    {
        "player_name" : ["권혁","이범호","필","박병호","양준혁"],
        "team_no" : [1993, 2001, 2001, 2010, None]
    },
    index = ["p_01","p-02","p_03","p_04","p_05"]
)
baseball_player.index.name="player_id"
baseball_player
```

```python
baseball_team1 = pd.DataFrame(
    {
        "team_name" : ["한화","두산","기아","넥센"],
        "team_home": ["대전","서울","광주","고척"]
    },
    index = [1993,1999,2001,2010]
)
baseball_team #팀 DF는 연결의 기준이 가로줄 인덱스
```

```python
#선수 : L, 팀 : R
#-> Left join (L기준 : 컬럼, R의 기준 : 가로 인덱스)
pd.merge(baseball_player1, baseball_team1, how = "left", left_on="team_no", right_index=True)
```

```python
#위와 동일한 상황에서 right join하면 
pd.merge(baseball_player1, baseball_team1, how = "right", left_on="team_no", right_index=True)
```

```python
# left exclusive join
# -> 1) left join을 먼저하고,
# -> 2) Right에서 오지 못한 정보에 대해서
```

```python
temp = pd.merge(baseball_player1, baseball_team1,
         how="left",
         left_on ="team_no", right_index = True)

temp[ temp["team_name"].isnull() ]
```

```python
# 오른쪽 팀 정보 빵구난 것
temp = pd.merge(baseball_player1, baseball_team1,
         how="right",
         left_on ="team_no", right_index = True)

temp[ temp["player_name"].isnull() ]
```

```python
# full
pd.merge(baseball_player1, baseball_team1,
         how="outer",
         left_on ="team_no", right_index = True)
```
