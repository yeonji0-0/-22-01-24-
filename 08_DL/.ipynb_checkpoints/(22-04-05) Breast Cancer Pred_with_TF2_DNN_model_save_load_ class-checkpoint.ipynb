#   데이터 분리 : Feature, Traget --> Target의 종류 및 수량 확인

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
np.unique(y)
```

```python
np.unique(y, return_counts=True)
```

# 단순 전처리!!

### 단순 전처리 --> target encoding

```python
# from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
y = enc.fit_transform(y)
np.unique(y, return_counts=True)
```

### 단순 전처리 --> train_test split

```python
# from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1234)
```

### 단순 전처리 --> standardScaler적용

```python
# from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Classfication Model

#### 아래는 앞에서 한 전통적인 ML 방법들을 가지고 분류모델을 하는 과정                       
#### -> 전처리, 데이터 분리 등 편한 패키지를 활용해서하면 된다
#### -> 원래는 각 모델별로 HypeerParameter 튜닝을 통해서 최적화 + Optuna
#### ++ 여러 모델을 확보를 하고, 이들에 대한 앙상블 ( voting, stacking)
#### --> 이제 활용하는 모델의 종류도 DL까지 확장이 되서 생각하면 된다

### 01) svm simple

```python
clf = SVC()

clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
print(accuracy)
```

### 02) knn simple

```python
clf = KNeighborsClassifier(n_neighbors = 3)
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
print(accuracy)
```

```python
## DL 기반의 모델을 만들어 보자
# 잠시) 분류 문제이기 때문에 출력에 대한 처리가 ML과 다르게 Ont-hot-enc방식으로 변환!!
```

* one-hot-encoding

```python
y_train_ohe  = tf.keras.utils.to_categorical(y_train, num_classes =2)
y_test_ohe  = tf.keras.utils.to_categorical(y_test, num_classes =2)
print(y_train_ohe[0])
print(y_test_ohe[0])
```

```python
X_train.shape
```

```python
dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
dataset_train = tf.data.Dataset.from_tensor_slices((X_train, y_train))
dataset_train
```

```python
for feat, targ in dataset_train.take(5):
    print ('Features: {}, Target: {}'.format(feat, targ))
```

```python
dataset_test = tf.data.Dataset.from_tensor_slices((X_test, y_test))
dataset_test
```

```python
X_train.shape
```

```python
dnn_model = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(input_shape = (30,), units = 256, activation = "relu"),
        tf.keras.layers.Dense(units = 128, activation = "relu"), 
        tf.keras.layers.Dense(units = 64, activation = "relu"),
        tf.keras.layers.Dense(units = 64, activation = "relu"),
        #출력용 outlayer 사용
        tf.keras.layers.Dense(units = 2, activation = "softmax")
    ]
)

dnn_model.compile(optimizer = "adam", loss="binary_crossentropy", metrics=["accuracy"])

# ML에서 사용한 정답지 y_train을 사용하면 안 됨
# 꼭 DL & 분류 -> 정답지 종류에 대한 one-hot-encoding로 바꿔서 넣어야 함
dnn_model.fit(X_train, y_train_ohe,epochs=15 )
```

```python
dnn_model.evaluate(X_test, y_test_ohe, verbose=2)
```

*** DNN을 활용하여 모델의 성능을 올리게 됨!!!

```python
dnn_model = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(input_shape = (30,), units = 256, activation = "relu"),
        tf.keras.layers.BatchNormalization(), # 히든 노드와 Batch정규화 시켜줌
        tf.keras.layers.Dropout(0.1), # 중간에 한 번씩 끊어줌
        tf.keras.layers.Dense(units = 128, activation = "relu"), 
        tf.keras.layers.Dropout(0.1),
        tf.keras.layers.Dense(units = 64, activation = "relu"),
        tf.keras.layers.Dropout(0.1),
        tf.keras.layers.Dense(units = 64, activation = "relu"),
        tf.keras.layers.Dropout(0.1),
        #출력용 outlayer 사용
        tf.keras.layers.Dense(units = 2, activation = "softmax")
    ]
)

dnn_model.compile(optimizer = "adam", loss="binary_crossentropy", metrics=["accuracy"])

# ML에서 사용한 정답지 y_train을 사용하면 안 됨
# 꼭 DL & 분류 -> 정답지 종류에 대한 one-hot-encoding로 바꿔서 넣어야 함
dnn_model.fit(X_train, y_train_ohe,epochs=15 )
```

```python
dnn_model.evaluate(X_test, y_test_ohe, verbose=2)
```

# 모델의 epoch를 하면서 중간 중간 모델의 weight를 저장을 하자!!!

* ref) https://www.tensorflow.org/tutorials/keras/save_and_load

### 01) epoch 할 때 마다 하나의 파일에 업데이트를 하면서 최종을 기록할 때.. --> 단, 불러서 사용할 때 모델은 정확하게 학습한 모델과 완전히!!!! 동일!!! 해야 함!!!!!  --> 주의!!!) model의 compile까지 해야함!!!!

```python
import os
```

```python
checkpoint_path = "C:/Users/MEDICI/Documents/MEDICI (22-01-19 ~ 22-06-24)/GitHub/08_DL/09_data/training_1/cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

# 체크포인트 콜백 만들기
cp_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)

#dnn_model.fit(X_train, y_train_ohe,epochs=15 ,callbacks = [cp_callback]) 
# dnn_model.fit(X_train, y_train_ohe,epochs=15 ,callbacks = [cp_callback]) 만 하게 되면...
# 참고로 위에서 학습하고 다시 또 돌리면 epoch가 다시 추가해서 돌리게 되므로 결과가 다르고, 더 돌리는 결과과 됨;;;
# 위에서 돌린 epoch +  지금 또 돌린 epoch 

# 그래서 다시 위에와 동일하게 하기 위해서..처음부터 다시...
dnn_model = tf.keras.Sequential([
    #tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu', input_shape=(30,)),
    #tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(128, activation='relu'),
    #tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(2, activation='softmax'),
])
dnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


dnn_model.fit(X_train, y_train_ohe,epochs=18 ,callbacks = [cp_callback])
```

```python
dnn_model.evaluate(X_test, y_test_ohe,verbose=2)
```

* 위에서 학습한 맨 마지막의 결과에 대한 가중치를 그대로 불러와서 사용을 한다!!! 단, 주의할 점은 compile까지 해야함!!!

```python
dnn_model_2 = tf.keras.Sequential([
    #tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu', input_shape=(30,)),
    #tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(128, activation='relu'),
    #tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(2, activation='softmax'),
])
# 주의!!! 모델 구조 + compile 까지 완전히 동일해야 함!!!!!
dnn_model_2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
dnn_model_2.load_weights(checkpoint_path)
dnn_model_2.evaluate(X_test, y_test_ohe, verbose=2)
```

## 매 epoch마다 저장을 해서, 필요한 때의 값을 가지고 와서 할 때...

```python
# 파일 이름에 에포크 번호를 포함시킵니다(`str.format` 포맷)
checkpoint_path = "C:/Users/MEDICI/Documents/MEDICI (22-01-19 ~ 22-06-24)/GitHub/08_DL/09_data/training_2/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)

cp_callback = tf.keras.callbacks.ModelCheckpoint(
    checkpoint_path, verbose=1, save_weights_only=True)



# 그래서 다시 위에와 동일하게 하기 위해서..처음부터 다시...
dnn_model = tf.keras.Sequential([
    #tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu', input_shape=(30,)),
    #tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(128, activation='relu'),
    #tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(64, activation='relu'), 
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(2, activation='softmax'),
    
])

dnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

###########
# 추가된 부분!!! --> 초기 값을 저장을 함...
dnn_model.save_weights(checkpoint_path.format(epoch=0))
##########

dnn_model.fit(X_train, y_train_ohe,epochs=20 ,callbacks = [cp_callback])

```

```python
dnn_model.evaluate(X_test, y_test_ohe,verbose=2)
```
