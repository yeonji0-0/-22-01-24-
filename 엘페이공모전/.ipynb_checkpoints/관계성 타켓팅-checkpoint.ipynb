```python
a = cus_304050.loc[((cus_304050['chnl_dv'] == 1) &( (cus_304050['de_hr'] >= 9) & (cus_304050['de_hr'] <= 12)) | ((cus_304050['de_hr'] >= 13) & (cus_304050['de_hr'] <= 18))) \
                   &((cus_304050['de_dt'].dt.dayofweek != 5) | (cus_304050['de_dt'].dt.dayofweek != 6)|(cus_304050['de_dt'] != '2021-01-01')|(cus_304050['de_dt'] != '2021-02-11')|(cus_304050['de_dt'] != '2021-02-12') \
                   |(cus_304050['de_dt'] != '2021-03-01')|(cus_304050['de_dt'] != '2021-05-05')|(cus_304050['de_dt'] != '2021-05-19')|(cus_304050['de_dt'] != '2021-09-20')|(cus_304050['de_dt'] != '2021-09-21')|(cus_304050['de_dt'] != '2021-09-22')), :]
a = cus_304050.groupby('cust').agg({'buy_ct': 'count'})
b = []
for i in range(len(a)):
    if a['buy_ct'][i] > 20: # 연차 수
           b.append(a.index[i])
```
