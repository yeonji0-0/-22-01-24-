### 시간 비교

```python
#필요한 패키지
import time
import pandas as pd
import numpy as np
from collections import deque
```

```python
#다양한 생성하는 방법에 대해서 이야기를 할 것임
#DF : 데이터를 합치는 방법 2-> append, concat : 빈번하게 여러번 하게 되면 concat

def df_append(n):
    df = pd.DataFrame(columns=["A","B","C"])
    temp = pd.DataFrame.from_dict({"A":[1], "B":[11.3], "C":["abcdef"]})
    for _ in range(n):
        df = df.append(temp, ignore_index = True)
    return df

def df_concat(n):
    df = pd.DataFrame(columns=["A","B","C"])
    temp = pd.DataFrame.from_dict({"A":[1], "B":[11.3], "C":["abcdef"]})
    for _ in range(n):
        df = pd.concat([df, temp], ignore_index = True)
    return df

def list_append_df(n):
    data = []
    for _ in range(n):
        data.append([1,11.3,"abcdef"])
    return pd.DataFrame(data, columns=["A","B","C"])

def np_append_df(n):
    data = np.array(np.empty((0,3), int))
    for _ in range(n):
        data = np.append(data, np.array([[1, 11.3, "abcdef"]]), axis =0)
    return pd.DataFrame(data, columns=["A","B","C"])

def deque_append_df(n):
    data = deque()
    for _ in range(n):
        data.append([1, 11.3, "abcdef"])
    return pd.DataFrame(data, columns=["A","B","C"])
```

```python
n = 10
%time df_append(n)
%time df_concat(n)
%time list_append_df(n)
%time np_append_df(n)
%time deque_append_df(n)
print("-"*10)
n = 1000
%time df_append(n)
%time df_concat(n)
%time list_append_df(n)
%time np_append_df(n)
%time deque_append_df(n)
print("-"*10)
n = 10000
%time df_append(n)
%time df_concat(n)
%time list_append_df(n)
%time np_append_df(n)
%time deque_append_df(n)
```
