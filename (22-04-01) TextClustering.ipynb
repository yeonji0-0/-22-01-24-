```python
# 리뷰 평점에 대해서 3점 미만 : negative 이상은 : postive로 변환 -> 함수
# 입력 : score 1~5점 받아서
# 출력  : Negative, Postive 문자열로
def det_pos_neg(x):
    if x < 3:
        return "Negative"
    elif x >= 3:
        return "Positive"
    else:
        return "None"

###### lambda 함수로 처리하기
# lambda x : (조건 만족할 때 값) if 조건 else (else 값...) # 2분법저그로 계속 처리
# lambda x : "Negative" if x < 3 else "Postive"
```

* text 전처리 : html 테그 지우기

```python
# html, xml 태그로 표현하는 것들을 찾는 패턴 : < .*?> -> html, xml
# 저번 데이터는 bs4로 지웠음
# 정규식으로..
i=0;
for sent in data['Text'].values:
    if (len(re.findall('<.*?>', sent))):
        print(i)
        print(sent)
        break;
    else:
        pass
    i += 1;
```

```python
# 영어 stemmer 중에서 snowball을 가지고 와서 사용 -> 한국어로 하시려면 koNLP 
snow_stem= nltk.stem.SnowballStemmer('english') 

# stopword에 대한 것을 가져와서 사용..
stop_words=set(stopwords.words('english'))

print(stop_words)
print("-"*100)
print('** Test **')

temp = "tasty"
print(snow_stem.stem(temp))
```

```python
# 전처리를 위한 간단한 함수 
# 1) html 태그 지우는 것 : 여기서는 간단한 정규식으로 지위
def cleanhtml(sentence):
    cleantag = re.compile('<.*?') # 정규식 표현을 명확하게 정의할 때 compile
    clean_text = re.sub(cleantag, " ", sentence)
    return clean_text

# 2) 정규식을 이용해서 문장부호에 대해서 제거
def cleanpunc(sentence):
    clean_text = re.sub(r"[?|!|~|#|.|,|(|)|/]", "", sentence) # -> 구글링해서 정규식.. + test + 수정해서
    return clean_text
```

```python
####### 리뷰들에 대해서 전처리를 하면서 
####### 긍정 리뷰들과 부정 리뷰들을 따로 분류해서 담으려고 한다~
t1 = time.time()
i=0
str1=' '
final_string=[]
all_positive_words=[] # 긍정 리뷰들을 담은 것
all_negative_words=[] # 부정 리뷰들을 담은 것
s=''

for sent in data['Text'].values:
    filtered_sentence=[] # 
    sent=cleanhtml(sent) # HTML의 테그들을 제거하는 함수들!!!!
    for w in sent.split(): # 문장을 공란으로 분리 --> 단어만다 돌아가면서..
        for cleaned_words in cleanpunc(w).split(): # 문장부호 제거하는 함수를 통하고 띄어쓰기 기준으로..
            if((cleaned_words.isalpha()) & (len(cleaned_words)>2) ): # 알파벳이면서 3알파벳.. 
                if(cleaned_words.lower() not in stop_words): #  stop은 아니여야 함!!!
                    s=(snow_stem.stem(cleaned_words.lower())).encode('utf8')
                    filtered_sentence.append(s) # 앞의 4가지 조건을 만족한 데이터만 밀어 넣음...
                    if (data['Score'].values)[i] == 'Positive': 
                        all_positive_words.append(s) # 필터  이후에 긍정인 것들만..
                    elif(data['Score'].values)[i] == 'Negative':
                        all_negative_words.append(s) #필터 이후 부정인 것들만.
                    else:
                        print("Check!!!!")
                else:
                    continue
            else:
                continue 
    str1 = b" ".join(filtered_sentence) # 1개의 글마다 다 정리..
    final_string.append(str1)
    i+=1
t2 = time.time()
print("Done!!" + str(t2-t1))
```

* silhouette_score : ![image.png](08_data/img/03.png)
군집화가 얼마나 잘 되었느냐에 대한 지표. 기본적인 개념은 얼마나 같은 군집에 대해서는 가깝고, 다른 군집에 대해서는 멀리 되어 있는지를 측정하는 지표임.a(i)는 같은 군집에 속한 다른 데이터들간의 평균적인 거리, b(i)는 i데이터에서 제일 가까운 다른 군집에 대한 평균적인 거리를 의미를 하고, a(i)-b(i)는 두 그룹간의 거리에 대한 지표임!!!
-1에 가까우면, 근처의 군집과 거리가 멀다
-0에 가까우면, 근처의 군집과 거리가 가깝다
-간혹 음수 , 잘못 되었다는 것;;;
--> 데이터의 수가 많으면 이 연산도 상당히 오래 걸린다;;;

```python
from sklearn import metrics
from sklearn.metrics import silhouette_samples, silhouette_score
# silhouette_samples : 모든 데이터들 각각의 실루엣 계수를 계산을 해서 반환을 한다.
# silhouette_score : 전체 데이터의 실루엣 계수값들의 평균값을 반환을 한다.
```

```python
# 상당히 시간이 걸림; -> full set 기준
silhouette_score = metrics.silhouette_score(bow, labels, metric="euclidean")
silhouette_score
```

* BoW, TF-IDF의 장점 : 문서의 유사도를 구하기가 쉽고 계산이 용이함, 관련이 있는 단어들의 수치값은 크게 유지가 되며, 빈도수는 높지만 여러 문서에 존재하는 단어는 낮추는 경향이 있음.<br>
* BoW, TF-IDF의 단점 : 단순 단어의 빈도 기반으로 되어 있어서, 의미 기반이 아닌 한계성이 있음, 동음이의어에는 상당히 취약함.
특히 BoW의 경우에는 의미가 비교적 없는 단어가 빈번하게 출현을 하면, 이것에 대한 가중치가 상당히 크게 적용이 될 수 있어서 문서 상에서의 중요한 단어의 특징을 나타내기에는 TF-IDF가 좀 더 효율적이라고 알려져 있음

## DBSCAN : Density-based Spatial Clustering of application with noise

![DBSCAN](08_data/img/10.png)

* DBSCAN은 k-means와 유사하지만, 처음에 군집의 수를 지정을 하지 않고, 자동적으로 최적의 군집의 갯수를 찾아가는 알고리즘임. -> 밀도를 활용해서 군집을 형성해 나가는 방식
* eps : 한 데이터가 주변에 얼마큰 떨어진 거리를 같은 군집으로 할지에 대한 거리기준
* min_sample은 적어도 한 군집에는 최소 이 갯수 이상의 모여야 군집으로 인정을 하겠다는 부분.
* 

```python
from sklearn.cluster import DBSCAN
```

```python
minPts = 2 * 100
# 참고) https://gist.github.com/m00nlight/0f9306b4d4e61ba0195f
def lower_bound(nums, target): 
    l, r = 0, len(nums) - 1
    while l <= r: # Binary searching.
        mid = int(l + (r - l) / 2)
        if nums[mid] >= target:
            r = mid - 1
        else:
            l = mid + 1
    return l

def compute200thnearestneighbour(x, data): 
    dists = []
    for val in data:
        dist = np.sum((x - val) **2 )
        if(len(dists) == 200 and dists[199] > dist): 
            l = int(lower_bound(dists, dist)) 
            if l < 200 and l >= 0 and dists[l] > dist:
                dists[l] = dist
        else:
            dists.append(dist)
            dists.sort()
    
    return dists[199] 
```

```python

twohundrethneigh = []
for val in sent_vectors[:1500]:
    twohundrethneigh.append( compute200thnearestneighbour(val, sent_vectors[:1500]) )
twohundrethneigh.sort()

```

```python
# Plotting for the Elbow Method :
plt.figure(figsize=(14,4))
plt.title("Elbow Method for Finding the right Eps hyperparameter")
plt.plot([x for x in range(len(twohundrethneigh))], twohundrethneigh)
plt.xlabel("Number of points")
plt.ylabel("Distance of 200th Nearest Neighbour")
plt.show()
```

```python
# 이것들을 바탕으로 사전을 구축해야 함!!!! -> 입력에 대한 리스트 형태 주의!!!!!!!!
dict_corpus = corpora.Dictionary(clean_corpus)
print(dict_corpus)
```

```python
# 그냥  dict_corpus를 하면 숫자만..아래와 같이 해야 실제 단어들을 볼 수 있음..
for i in dict_corpus.values():
    print(i)
```

```python
# 정제된 단어리스트들을 하나씩 뽑아서..사전에 구축한 딕션어리에 맞춰서 카운팅을.....
doc_term_matrix = [ dict_corpus.doc2bow(i) for i in clean_corpus]
doc_term_matrix
```

```python
# Topic Model을 위해서는 문서별로 내가 수집한 문서에 대한 단어들의 숫자들에 대한 행렬로 변환!!!!!!
lda = gensim.models.ldamodel.LdaModel
```

```python
# 주어진 문서를 도대체 몇 개의 주제로 나눌꺼냐???
n = 10 

# 입력으로 가장 중요한 문서별 단어 숫자로 표현한  doc-to-term_matrix
ladmodel = lda(doc_term_matrix, num_topics=n, id2word=dict_corpus, random_state=123)
```
