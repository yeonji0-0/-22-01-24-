## Pivot

```python
import numpy as np
import pandas as pd
# -> colab에서 하기에 파일 불러오기 위해서 
#from google.colab import files
#myfiles = files.upload()
```

```python
data = pd.read_excel("01_data/11_sales-funnel.xlsx")
data.head()
```

```python
#설정 : 영업관련 데이터 셋인데 -> 영업팀에 대한 인사평가 진행
#잠시) 카테고리 변수로 명확하게 선정하는 부분 : pandas에서 
data["Status"].unique()
```

```python
data["Status"] = data["Status"].astype("category")
data.head()  #데이터에서는 차이 없지만 아래 유니크 다시 해보면 달라짐
```

```python
data["Status"].unique()
```

```python
# -> 특별하게 순서에 대해서 지정을 하지 않으면 알파벳 순서
#-> 내가 원하는 순서대로 : 금, 은 ,동
data["Status"].cat.set_categories(["won","pending","presented","declined"], inplace = True)
data
```

```python
data["Name"].unique()
```

```python
#가로줄에 대한 세팅 : index
#내가 고객별로 볼 수 있는 항목 그냥 보여줘
# --> 그래서 불필요한 컬럼도 나오게 됨
pd.pivot_table(data, index=["Name"])
```

```python
#가로줄에 보고자 한느 기준이 여러개일 수 있음
pd.pivot_table(data, index = ["Name","Rep"])
```

```python
#인사팀이라고 하면, 영업팀장에 2명이 있어서
# --> 이 중에서 누구를 영업팀 임원으로 선발할까?
# ---> 보고 싶은 타겟 : 영업팀장 - 영업사원 기준으로 보자
pd.pivot_table(data, index = ["Manager","Rep"])
```

```python
#이 사람들의 판매한 가격을 보고자 한다 (세로줄)
pd.pivot_table(data, index = ["Manager","Rep"], values = ["Price"])
```

```python
#이러한 경우들은 평균매출보다는 -> 전체 총 매출, 영업실적으로 채우려고 함 (평균X)
#-> 총합으로 채우려고 함
pd.pivot_table(data, index = ["Manager","Rep"], 
                     values = ["Price"], 
                     aggfunc = [np.sum])
```

```python
pd.pivot_table(data, index = ["Manager","Rep"], 
                     values = ["Price"], 
                     aggfunc = [np.mean, len])
```

```python
# 가격을,,,,,상품별로 나눠서 보고 싶다
pd.pivot_table(data, index =["Manager","Rep"],
                     values = ["Price"],
                    columns =["Product"],
                    aggfunc = [np.sum] )
```

```python
# 가격을,,,,,상품별로 나눠서 보고 싶다
# -> 너무 많이 조건별로 쪼개면 빵구날 수 있음!!
pd.pivot_table(data, index =["Manager","Rep"],
                     values = ["Price"],
                    columns =["Product"],
                    aggfunc = [np.sum],
                    fill_value= 0)
```

```python
# 가격을,,,,,상품별로 나눠서 보고 싶다
# -> 너무 많이 조건별로 쪼개면 빵구날 수 있음!!
pd.pivot_table(data, index =["Manager","Rep"],
                     values = ["Price","Quantity"],
                    columns =["Product"],
                    aggfunc = [np.sum],
                    fill_value= 0)
```

```python
# 가로 : 팀장(Manager), 영업사원(Rep), 상품(product)
# 세로 : price, quantity 
# 채움 : 합
pd.pivot_table(data, index = ["Manager","Rep","Product"],
                    values = ["Price","Quantity"],
                    aggfunc = [np.sum],
               fill_value=0 )
```

```python
pd.pivot_table(data, index = ["Manager","Rep","Product"],
                    values = ["Price","Quantity"],
                    aggfunc = [np.sum],
               fill_value=0,
               margins=True)
```

```python
# 각 세로줄에 대해서 각기 다른 연산을 보고자 한다
# -> dict를 통해서 각기 보고자 하는 세로줄values
#    각각 무엇을 연산을 시켜서 볼지 k=v로 지정
pd.pivot_table(data, index =["Manager","Status"],
                     values = ["Price","Quantity"],
                    columns =["Product"],
                    aggfunc = {"Price":[np.sum, np.mean],
                               "Quantity":len},
                    fill_value= 0)
```

```python
# pivot_table하고 쿼리문처럼 조건 검색
table = pd.pivot_table(data, index =["Manager","Status"],
                     values = ["Price","Quantity"],
                    columns =["Product"],
                    aggfunc = {"Price":[np.sum, np.mean],
                               "Quantity":len},
                    fill_value= 0)
table.query('Manager=="Debra Henley"') # 쿼리문을 조건에 대해서 문자열로
```

```python
#특별한 케이스로 주로 이렇게 만들면서
#셀을 단순 카운팅!! 몇 건 있었는지에 대한 'crosstab' : 가로, 세로 -> 무조건 안은 빈도로 채운다
pd.crosstab(data["Name"],data["Price"])
```

```python
#pd.crosstab([가로],[세로])
pd.crosstab([data["Manager"],data["Rep"]],
            [data["Price"],data["Quantity"]])
```

```python
# 필요한 패키지 불러오기
# 내가 주피터노브북에서 실행한 그래프를 아래 출력창에 나타낼 때
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
```

> 참고) 아래 옵션을 수행하면 아래에 그래프가 나타나고, 아니면 창으로 나타남!!    :   %matplotlib inline

```python
# y = X^2의 그래프 생성을 위해서. --> 10개 구간으로 하기 위해서는 11로 처리해야 함! 구간이므로
x = np.linspace(start = 0, stop = 10, num=11)
y = x ** 2

print(x)
print(y)
```

```python
#01) 그래프의 위치 지정해서 그리는 방법

# 그릴 준비 : 스케치북 준비
fig = plt.figure()

# 그림을 위치할 위치에 대한 포지션 지정
#axes = fig.add_axes([0.5, 0.5, 0.5, 0.5]) # left, bottom, width, height [0 ~ 1]
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height [0 ~ 1]


# 실제 데이터를 표기 - X데이터, y데이터, 그릴 옵션 [여기서는 red로 그리겠다는 의미]
axes.plot(x,y,"r")

# 기타 옵션 표시
# 01) x축 이름
axes.set_xlabel("x")
# 02) y축 이름
axes.set_ylabel("y")
# 03) 그래프 제목
axes.set_title("Test Graph")
```

```python
#  안에 그래프 넣기
fig = plt.figure()

# 메인 그래프와 안에 그릴 작은 그래프에 대한 위치 설정
# 아래의 위치에 대한 정보를 변경하면, 각기 그래프들의 위치가 변경됨
axes1 = fig.add_axes([0, 0,1,1])# 메인 그래프 위치
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3]) # 안에 그릴 그래프 위치


# 메인 그래프
axes1.plot(x,y,"r")
axes1.set_xlabel("x")
axes1.set_ylabel("y")
axes1.set_title("Title`")

# 안에 그릴 그래프
axes2.plot(y,x,"g")
axes2.set_xlabel("y")
axes2.set_ylabel("x")
axes2.set_title("Invert Title")
```

```python
# 02) 그래프의 위치 지정하지 않고, 분할해서 그리는 방법

# 분할을 1개로 처리
fig, axes = plt.subplots()

axes.plot(x,y,"red")

# 01) x축 이름
axes.set_xlabel("x")
# 02) y축 이름
axes.set_ylabel("y")
# 03) 그래프 제목
axes.set_title("Test Graph")
```

```python
# 기본적인 scatter / step / bar / fill_between
fig, axes = plt.subplots(1, 4, figsize=(20,5))

axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.3, alpha=0.1)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**2+10, color="blue", alpha=0.3);
axes[3].set_title("fill_between");
```
