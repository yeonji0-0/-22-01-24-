```python
# 1)반복에 대한 부분을 재귀함수로 구현 (규칙 그대로 함수화)
def fibo(x):
    if x == 1 or x == 2:
        return 1
    else:
        return fibo(x - 1) + fibo(x - 2)
print(fibo(5))
print(fibo(10))
```

```python
import time
```

```python
# 2)재귀함수 구현에 대한 시간 체크(재귀함수는 속도에 대한 이슈 있음)
t1 = time.time()
print(fibo(10))
print(time.time()-t1)
##############################
t1 = time.time()
print(fibo(20))
print(time.time()-t1)
##############################
t1 = time.time()
print(fibo(40))
print(time.time()-t1)
```

```python
#밑에서부터 : 원하는게 fibo(n)
#재귀함수는 fibo(n) -> fibo(n-1) : top down
#1,2,3,4, --> n 밑에서부터 위로 : bottom up
d = [0] * 1000
d[1] = 1
d[2] =1
n = 80
for i in range(3, n+1): #파이썬의 특징- 초기조건에 따라 세팅
    d[i] = d[i-1] + d[i-2]
print(d[n])
```
