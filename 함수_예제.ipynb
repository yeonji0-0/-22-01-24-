```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits import mplot3d
import matplotlib.font_manager as mfm

# numpy 출력 형식 지정
np.set_printoptions(precision=4, linewidth=150)

# matplotlib 스타일 지정
mpl.style.use('bmh')
mpl.style.use('seaborn-whitegrid')
style = plt.style.library['bmh']
# 스타일 컬러를 쉽게 쓸 수 있도록 리스트 저장
style_colors = [ c['color'] for c in style['axes.prop_cycle'] ]

# 그림을 로컬 폴더에 저장하고 싶으면 True로 수정 
file_print = False
```

```python
fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.set_zlabel('$z$', fontsize=20)

######################################
# 입력으로 할 t의 값들..
t = np.linspace(0, 2, 101)
#########################################
# 입력에 따른 출력들에 대한 것
x = np.sin(3*t)
y = np.cos(3*t)
z = t**2 / 2
########################################
# 위의 출력의 차원이 3차원이므로 3차원으로 표현하기 위해서.
ax.plot3D(x, y, z, c='k')
###########################################
ax.plot([x[0]],  [y[0]],  [z[0]],  'o', markersize=10, color='k',
        label="t = {:.2f}".format(t[0]))
ax.plot([x[50]], [y[50]], [z[50]], '^', markersize=10, color='k',
        label="t = {:.2f}".format(t[50]))
ax.plot([x[-1]], [y[-1]], [z[-1]], '*', markersize=10, color='k',
        label="t = {:.2f}".format(t[-1])) 

ax.legend(fontsize=15, loc="upper left")

plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(1, 1, 1)
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)

x1 = np.linspace(-3,3, 50)
x2 = np.linspace(-1, 5, 50)
X1, X2 = np.meshgrid(x1, x2)
Z = 80*(X2 - X1**2)**2 + (6-X1)**2

# 3가지의 높이에 대해서 r,g,b로 그려본다. 
# 여기서 각 높이별로 여러 부분으로 나타날 수 있으니, 이것의 나오는 갯수별로 보고  
# 아래 그래프를  그려나가야 한다.
contours = ax.contour(X1, X2, Z, levels=[100,300,500],  colors=['r','g','b'], 
            linestyles=['solid','dashed', 'dotted'])

ax.clabel(contours, contours.levels, fmt="%d", inline=True, fontsize=20)

ax.set_xlabel(r'$x_1$', fontsize=20)
ax.set_ylabel(r'$x_2$', fontsize=20)

plt.show()
```

```python
import numpy as np
```

```python
x = tf.Variable(np.array([ 
                 [1], 
                 [2] , 
                 [3]
                 ]), dtype=tf.float32)
x
```

```python
xT = tf.transpose(x)
xT
```

```python
x_2norm_matmul = tf.math.sqrt(tf.matmul(xT,x))
x_2norm_matmul
```

```python
x_2norm = tf.norm(x, ord=2)
x_2norm
```
